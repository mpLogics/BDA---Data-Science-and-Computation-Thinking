# The Stock Market : Moderating Stock Price on the basis of a fair coin toss
The simulation setup, results, analysis and all other necessary details are mentioned in the document in the same folder.

```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def stockPredict(Days,NumberofStocks,BasePrice):
    price = np.zeros(NumberofStocks)
    Mean = np.zeros(NumberofStocks)
    Variance = np.zeros(NumberofStocks)
    for i in range(NumberofStocks):
        price[i]=BasePrice
    Rate = np.zeros((NumberofStocks,Days))
    for j in range(NumberofStocks):
        for i in range(Days):
        
            if random.random() < 0.5:
                price[j] = price[j] + 1
                Rate[j][i] = price[j]
        
            elif price[j]==0:
                continue
            else:
                price[j] = price[j] - 1
                Rate[j][i] = price[j]

    plt.figure(figsize=(15,8))
    for i in range(NumberofStocks):
        plt.plot(Rate[i],label = "Stock "+(str)(i+1))
        if NumberofStocks<10:
            plt.legend()
        plt.xlabel("Days")
        plt.ylabel("Stock Price")
    plt.title("Daily Stock Price Value for 10k days")
    plt.show()
    for i in range(NumberofStocks):
        Mean[i]=np.mean(Rate[i])
        Variance[i]=(np.var(Rate[i]))
        
    
    plt.figure(figsize=(15,8))
    plt.plot(Mean,label="Expected Value")
    plt.plot(Variance,label="Variance")
    plt.title("Expected Value and Variance for the stock prices")
    plt.legend()
```

```python
Days = 10000
NumberofStocks = 100
BasePrice = 100
stockPredict(Days,NumberofStocks,BasePrice)
```
