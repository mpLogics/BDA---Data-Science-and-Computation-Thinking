# Monte Carlo simulation for finding the MLE for Î» given r = 9

The detailed methodology, simulation setup, analysis and takeaways can be found in the document along with this code. It has both the mathematical as well as simulation results. All the distributions have also been generated for a certain number of iterations and then results have been observed again.

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def PoissonCalculator(lamda, x):
    return (math.exp(-lamda)*(lamda**x))/math.factorial(x)
```

```python
def distributionGeneration(size):
    Dist1 = np.zeros(size)
    Dist2 = np.zeros(size)
    Dist3 = np.zeros(size)
    sum = 0
    for i in range((int)(size/2)):
        sum = sum + 0.5
        Dist1[i] = sum
    for i in range(size):
        Dist2[i] = random.uniform(0,size)
        Dist3[i] = random.randint(0,size)
    
    return Dist1, Dist2, Dist3        
```

```python
def MonteCarlo(Size,R,b):
    D1 = np.zeros(Size)
    D2 = np.zeros(Size)
    D3 = np.zeros(Size)

    D1, D2, D3 = distributionGeneration(Size)

    Values1 = np.zeros(Size) # For distribution 1
    Values2 = np.zeros(Size) # For distribution 2
    Values3 = np.zeros(Size) # For distribution 3

    for i in range(Size):
        Values1[i] = PoissonCalculator(D1[i] + b,R)
        Values2[i] = PoissonCalculator(D2[i] + b,R)
        Values3[i] = PoissonCalculator(D3[i] + b,R)
    
    #Values[i] = PoissonCalculator(i+1,9)    
    
    plt.scatter(D1,Values1,label = "Max lambda value: " + (str)(D1[np.argmax(Values1)]))
    plt.xlabel("Lambda Value")
    plt.ylabel("Likelihood value for r = 9")
    print("Maximum value for function: " + (str)(max(Values1)) + " at lambda = " + (str)(D1[np.argmax(Values1)]))
    print("\n")
    plt.legend()

    print("Using other distributions:")

    print("Max value for lambda: ",D2[np.argmax(Values2)])
    print("Function value for given lambda: ",max(Values2))

    print("\nMax value for lambda: ",D3[np.argmax(Values3)])
    print("Function value for given lambda: ",max(Values3))

    #print(D3[np.argmax(Values3)],max(Values3))
    plt.show()
    
    return D1[np.argmax(Values1)],D2[np.argmax(Values2)],D3[np.argmax(Values3)],max(Values1),max(Values2),max(Values3)
```

```python
Size = 1000
R = 9
b = 0
result1 = np.zeros(3)
result2 = np.zeros(3)
maxValue = 0
maxLambda = 0
for i in range(100):
    result1[0],result1[1],result1[2],result2[0],result2[1],result2[2] = MonteCarlo(Size,R,b)
    if maxValue < max(result2):
        maxValue = max(result2)
        maxLamda = result1[np.argmax(result2)]
    
print("The maximum possible value for Lambda: ",maxLamda)

```
