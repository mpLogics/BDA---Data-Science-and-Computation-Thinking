This function calculates the p-norm for a vector V.

```python
def vectorNorm(v,p):
  norm = 0.0;
  i = len(v)
  while i>0:
    norm = norm + v[i-1]**p
    i=i-1
  norm = norm **(1/p)
  return norm
```

pAnalyze allows us to see the trend of the values for norm for different values of p for a fixed 2-d vector. We notice that as p tends to 0, vector norm for the given vector tends to a high value. For p = 0, it will not be defined. Thus, we will have to see in the subsequent sections what might be a potential solution.

```python
def pAnalyze(pMax,changeSize,pMin):
  v = np.array([3,4])
  p1 = pMax
  p = pMin
  i = 0
  x1 = np.empty((int)((pMax-pMin)/changeSize))
  y1 = np.empty((int)((pMax-pMin)/changeSize))
  while p<pMax:
    y1[i] = vectorNorm(v,p)
    x1[i] = p
    p=p+changeSize
    i=i+1

  plt.plot(x1,y1)
  plt.grid(alpha =.6, linestyle ='--') 
  plt.title("Plot for " + str(pMin) + " < p < " + str(pMax))
  plt.xlabel("Values for P")
  plt.ylabel("Norm values")
  plt.show()
```

```python
pAnalyze(0.5,0.005,0.1)
print("\n")
pAnalyze(1,0.005,0.5)
print("\n")
pAnalyze(80,0.5,1)
```

This section forms a unit ball for p = 2. The graph of the ball can be seen uponn executing the code

```python
total = 100
p = 2
pointsX = []
pointsY = []
for i in range(total):
  for j in range(total):
    if vectorNorm([(i+1)/total,(j+1)/total],p)<= 1:
      pointsX.append(i)
      pointsY.append(j)


plt.scatter(pointsX,pointsY,c = 'lightblue')

pointsX = [-x for x in pointsX]
plt.scatter(pointsX,pointsY,c = 'lightblue')

pointsX = [-x for x in pointsX]
pointsY = [-x for x in pointsY]
plt.scatter(pointsX,pointsY,c = 'lightblue')

pointsX = [-x for x in pointsX]
plt.scatter(pointsX,pointsY,c = 'lightblue')
plt.figure(figsize=(100,80))
plt.show()
```
